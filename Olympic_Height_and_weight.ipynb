```python
def convert_to_mean(value):
    if '-' in value:
        min_val, max_val = map(int, value.split('-'))
        return str((min_val + max_val) / 2)
    elif ',' in value:
        min_val, max_val=map(int,value.split(','))
        return str((min_val + max_val) / 2)
    else:
        return value
```

```python
g1['weight']=g1['weight'].apply(convert_to_mean)
s1['weight']=s1['weight'].apply(convert_to_mean)
b1['weight']=b1['weight'].apply(convert_to_mean)
```
